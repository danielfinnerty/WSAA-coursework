Import required packages

```python
from github import Github, Auth 
import requests
from config import apikeys as cfg
```

`Auth` requires importing due to PyGithub now requiring authentication ([source](https://pygithub.readthedocs.io/en/stable/examples/Authentication.html)).
`from config import apikeys as cfg` directs to the config file, in which the API key resides.

```python
# specify API key 
apikey = cfg["githubkey"]

# Authenticate with PyGithub
auth = Auth.Token(apikey)
g = Github(auth=auth)

# Specify repository
repo = g.get_repo("danielfinnerty/WSAA-Assignment04-Private")

# Confirm repository
print(repo.clone_url)
```

With the API key authenticated, and the repository location confirmed, the file can now be downloaded, modified, and pushed back to the repository.

```python
# Specify file within the repository
filename = repo.get_contents("assignment04.txt")

# Specify file URL
urloffile = filename.download_url

# Show file path if required
# print (urloffile)

# Download file and show content
response = requests.get(urloffile)
filecontent = response.text
print(f"The current file content is:\n" + filecontent)
```

```python
# Replace data in file; Andrew replaced by Daniel
newfiletext = filecontent.replace("Andrew", "Daniel") #https://www.w3schools.com/python/ref_string_replace.asp

# Show new file test
print(f"The new file content is:\n" + newfiletext)

# Upload and commit to github
gitHubResponse=repo.update_file(filename.path,"updated by assignment04 code",newfiletext,filename.sha)

# Show github response to very it succeeded.
print (gitHubResponse)
```
